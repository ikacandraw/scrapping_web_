```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
#key = input('please enter the term :')
#location = input('please enter the location too :')
url = 'https://www.simplyhired.com/search?q=++&l='
headers = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36'
}
```

```python
req = requests.get(url,headers=headers)
```

```python
soup = BeautifulSoup(req.text,'html.parser')
items = soup.findAll('ul', {'id':"job-list"})
print(items)
datas = []
for it in items:
    name = it.find('a', class_="chakra-button css-12bkbc3").text
    company = it.find('span', {'data-testid':"companyName"}).text
    area = it.find('span', {'data-testid':"searchSerpJobLocation"}).text
    fee = ''
    try:
     fee = it.find('p', class_="chakra-text css-1ejkpji").text
    except:
      continue
    qualification =''
    #try:
      #qualification = it.find('ul', class_="chakra-wrap__list css-wah4g8").text
    #except:
      #pass
    datas.append([name,company,area,fee])
```

```python
datas = []
count_page = 0
for page in range(1,5):
    count_page+=1
    print('scraping page:',count_page)
    req = requests.get(url,headers=headers)
    soup = BeautifulSoup(req.text,'html.parser')
    items = soup.findAll('li', class_="css-0")
    for it in items:
        name = it.find('a', class_="chakra-button css-12bkbc3").text
        company = it.find('span', {'data-testid':"companyName"}).text
        area = it.find('span', {'data-testid':"searchSerpJobLocation"}).text
        fee = ''
        try:
          fee = it.find('p', class_="chakra-text css-1ejkpji").text
        except:
          continue
        #qualification = it.find('ul', class_="chakra-wrap__list css-wah4g8").text
        datas.append([name,company,area,fee])
    items2 = soup.findAll('div',class_="css-1vp1jvg")
    for itemm in items2:
        qualification = ''
        try:
          qualification = itemm.find('ul', class_="chakra-wrap__list css-wah4g8").text
        except:
          pass
        datas.append([qualification])
```

```python
datas
```

```python
## Store data
head = ["Job Name","Company","Area","Fee","Qualification"]

with open('result.csv','w') as f:
  writer = csv.writer(f)
  # write the header
  writer.writerow(head)
  for data in datas:
    writer.writerow(data)

# writer = csv.writer(open("result.csv",'w',newline=''))
# writer.writerow(head)
# for d in datas: writer.writerow(d)
```
